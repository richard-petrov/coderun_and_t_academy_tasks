## Типы данных

```python
# 1. Числа (int, float)
возраст = 30
цена = 99.99
print(f"Целое: {возраст}, Дробное: {цена}, Сумма: {возраст + цена}")

# ---
# 2. Строки (str)
имя = "Алексей"
приветствие = 'Привет!'
print(f"Имя: {имя}, Длина: {len(имя)}, Первая буква: {имя[0]}")

# ---
# 3. Списки (list)
фрукты = ["яблоко", "банан", "апельсин"]
фрукты.append("манго")
print(f"Список: {фрукты}, Первый: {фрукты[0]}, Изменён: {фрукты[1]}") # банана нет, теперь вишня или что-то другое после изменений

# ---
# 4. Кортежи (tuple)
координаты = (10, 20)
print(f"Кортеж: {координаты}, X: {координаты[0]}")
# координаты[0] = 30 # Это вызовет ошибку, кортежи неизменяемы!

# ---
# 5. Множества (set)
цвета = {"красный", "синий", "красный"}
цвета.add("зеленый")
print(f"Множество: {цвета}, Уникальные элементы, есть ли синий? {'синий' in цвета}")

# ---
# 6. Словари (dict)
профиль = {"имя": "Анна", "возраст": 25}
профиль["город"] = "Москва"
print(f"Словарь: {профиль}, Имя: {профиль['имя']}, Город: {профиль['город']}")

# ---
# 7. Булевы значения (bool)
статус = True
результат = (10 > 5)
print(f"Статус: {статус}, 10 > 5: {результат}")
```

## Операторы

```python
# Пример 1: Простейшее условие if
возраст = 20

# Если возраст больше или равен 18, выводим сообщение
if возраст >= 18:
    print("Вам разрешено голосовать.") # Этот код выполнится, если условие True

# ---

# Пример 2: if-else (Если... Иначе...)
температура = 28

# Если температура выше 25, говорим "жарко", иначе - "нормально"
if температура > 25:
    print("На улице жарко.")
else:
    print("На улице нормальная температура.")

# Изменим температуру, чтобы увидеть другой результат
температура = 20
if температура > 25:
    print("На улице жарко.")
else:
    print("На улице нормальная температура.") # Теперь выполнится этот блок

# ---

# Пример 3: if-elif-else (Если... Иначе если... Иначе...)
# Это позволяет проверять несколько условий последовательно
оценка = 75

if оценка >= 90:
    print("Отлично!")
elif оценка >= 70: # Проверяется, если первое условие (оценка >= 90) было False
    print("Хорошо.")
elif оценка >= 50: # Проверяется, если предыдущие условия были False
    print("Удовлетворительно.")
else: # Выполняется, если ни одно из предыдущих условий не было True
    print("Неудовлетворительно.")

# ---

# Пример 4: Использование логических операторов в условиях
день_недели = "Суббота"
время = 14

# Условие: если это выходной И время после 12, то можно отдыхать
if день_недели == "Суббота" or день_недели == "Воскресенье": # Проверяем, выходной ли
    if время > 12: # Вложенное условие
        print("Можно расслабиться, выходной и время после полудня!")
    else:
        print("Выходной, но еще не совсем обед.")
else:
    print("Будни, нужно работать.")

# Можно написать это и в одну строку с помощью 'and'
# if (день_недели == "Суббота" or день_недели == "Воскресенье") and время > 12:
#     print("Можно расслабиться, выходной и время после полудня!")
# else:
#     print("Либо не выходной, либо еще рано.")

# ---

# Пример 5: Проверка наличия элемента в списке
продукты = ["молоко", "хлеб", "сыр"]
что_ищем = "хлеб"

if что_ищем in продукты: # Оператор 'in' проверяет наличие элемента
    print(f"'{что_ищем}' есть в списке покупок.")
else:
    print(f"'{что_ищем}' нет в списке покупок.")

что_ищем = "яйца"
if что_ищем in продукты:
    print(f"'{что_ищем}' есть в списке покупок.")
else:
    print(f"'{что_ищем}' нет в списке покупок.")
```

## while и for

```python
# Пример while: отсчет до 5
счетчик = 1
print("Начинаем отсчет с While:")
while счетчик <= 5: # Пока счетчик меньше или равен 5
    print(счетчик)
    счетчик += 1 # Увеличиваем счетчик, чтобы не было бесконечного цикла
print("Отсчет завершен!")
```

```python
# Пример for: перебор списка фруктов
фрукты = ["яблоко", "банан", "вишня"]
print("\nПеребираем фрукты с For:")
for фрукт in фрукты: # Для каждого фрукта в списке 'фрукты'
    print(фрукт)

# Пример for: перебор чисел в диапазоне (от 0 до 4)
print("\nПеребираем числа с For и range():")
for число in range(5): # range(5) генерирует числа 0, 1, 2, 3, 4
    print(число)

# Пример for: перебор строки
слово = "Python"
print("\nПеребираем буквы слова с For:")
for буква in слово:
    print(буква)
```

## Функции

```python
# Определение простой функции, которая ничего не принимает и ничего не возвращает
def поприветствовать():
    """
    Эта функция просто выводит приветствие.
    """
    print("Привет! Добро пожаловать в мир функций!")

# Вызов функции (запуск кода внутри неё)
поприветствовать() # Теперь эта строка выполнит код внутри поприветствовать()
поприветствовать() # Можем вызвать её ещё раз

# ---

# Функция, которая принимает аргумент (входные данные)
def поприветствовать_по_имени(имя):
    """
    Эта функция принимает имя и выводит персонализированное приветствие.
    :param имя: Имя человека для приветствия.
    """
    print(f"Привет, {имя}! Рад тебя видеть.")

# Вызов функции с аргументом
поприветствовать_по_имени("Алексей")
поприветствовать_по_имени("Анна")

# ---

# Функция, которая возвращает значение
def сложить_числа(а, б):
    """
    Эта функция складывает два числа и возвращает их сумму.
    :param а: Первое число.
    :param б: Второе число.
    :return: Сумма двух чисел.
    """
    сумма = а + б
    return сумма # Ключевое слово 'return' отправляет значение обратно

# Вызов функции, сохраняя возвращаемое значение в переменную
результат_сложения = сложить_числа(5, 3)
print(f"Результат сложения: {результат_сложения}") # Выведет 8

# Можно использовать возвращаемое значение сразу
print(f"2 + 7 = {сложить_числа(2, 7)}") # Выведет 9

# ---

# Функция с аргументами по умолчанию
def приветствие_с_городом(имя, город="Неизвестный город"):
    """
    Приветствует человека, указывая его город. Если город не указан, используется значение по умолчанию.
    :param имя: Имя человека.
    :param город: Город человека (по умолчанию "Неизвестный город").
    """
    print(f"Привет, {имя} из {город}!")

приветствие_с_городом("Мария", "Лондон") # Указываем город
приветствие_с_городом("Игорь")           # Используем город по умолчанию
```

## Работа с файлами

```python
# запись

with open("../../../data/test/test.txt", "w", encoding="UTF-8") as file:
    file.write("Hello, world!\n")
    file.write("Это тестовая запись в файл.\n")
```

```python
# чтение

with open("../../../data/test/test.txt", "r", encoding="UTF-8") as file:
    content = file.read()
    print(content)  # Выводим содержимое файла на экран
```

```python
# чтение построчно

with open("../../../data/test/test.txt", "r", encoding="UTF-8") as file:
    for line in file:
        print(line.strip())
```

```python
# добавление 

with open("../../../data/test/test.txt", "a", encoding="UTF-8") as file:
    file.write("Это добавочная строка!\n")
```

## Обработка исключений

```python
try:
    res = 10 / 0
except ZeroDivisionError as e:
    print(f"Ошибка: {e}")
except Exception as e:
    print(f"Произошла другая ошибка: {e}")
```
